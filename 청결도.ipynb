```python
# 서울시 생활안전도 통계(2011~2018)
# 2018 데이터 없음.

import pandas as pd
clean = pd.read_csv('서울시 생활안전도 통계.txt', sep='\t')
clean = clean[clean.columns[[0,1,5,9,13]]]
mask_1 = clean['대분류'].isin(['대분류', '서울시', '성별', '연령별', '학력별', '주택형태', '입주형태', '가구원수', '소득',
       '지역대분류'])
clean = clean[~mask_1]
mask_2 = clean['기간'].isin(['2018'])
clean = clean[~mask_2]
mask_3 = clean['대기오염'].isin(['…'])
clean = clean[~mask_3].reset_index(drop=True)
clean = clean.rename(columns={'대분류':'자치구', '길거리의 쓰레기방치':'길거리_쓰레기방치'})
clean = clean.astype({'대기오염':float, '수질오염':float, '길거리_쓰레기방치':float})
clean
```

```python
gu = clean['자치구'].unique()

air_mean=list()  # b1_mean=[]
water_mean=list()
trash_mean=list()

for i in range(len(gu)):  # python이 권장하지 않는 방식.
    a = gu[i]  # one of 자치구들, python이 권장하지 않는 방식.
    b = clean[clean['자치구']==a]  # 그 자치구에 해당하는 rows
    b1_mean = b['대기오염'].values.mean()  # np.array이므로 .mean() 가능
    b2_mean = b['수질오염'].values.mean()  # clean.columns[3]
    b3_mean = b['길거리_쓰레기방치'].values.mean()  # dtype('float64')
    air_mean.append(b1_mean)
    water_mean.append(b2_mean)
    trash_mean.append(b3_mean)
    
clean_mean = pd.DataFrame({'자치구':gu, '대기오염':air_mean, '수질오염':water_mean, '길거리_쓰레기방치':trash_mean})
clean_mean

# 'NoneType' object has no attribute 'append'
# air_mean = air_mean.append(b1_mean)하면 안되고, 그냥 air_mean.append(b1_mean)해줘야 함.
# You are not supposed to assign it to any variable, when you append something in the list, it updates automatically. use only:-
# if you assign this to a variable "last_list" again, it will no more be a list (will become a none type variable since you haven't declared the type for that) and append will become invalid in the next run.

# IndentationError: unexpected indent 
# for문 밑으로는 1단계 수준의 동일 tab line에 있어야 함.

```
