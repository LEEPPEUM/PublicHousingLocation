```python
# nan이 생기면서 컬럼이 제대로 바로 옆에 안붙는 현상은 해결함.
# 그러나 a_1, a_2, b의 컬럼 '자치구' 값의 순서가 다르므로 그냥 concat을 할 경우, 데이터의 배열 순서가 맞지 않는 모습.

a_1 = a_1.reset_index(drop=True)
a_2 = a_2.reset_index(drop=True)
b = b.reset_index(drop=True)

c = pd.concat([a_1,a_2,b], axis=1)
c

# df = pd.concat(ignore_index=True)
# concat의 파라미터

# df.reset_index(drop=True)
# function임.

# 둘다 같은 효과.
```

```python
# merge는 2개씩만 가능.
# on = '' / primary key(pk)와 같이 id라는 컬럼을 공통으로 가지고 있는 데이터라면 id라는 키로 매칭하여 데이터를 붙이는 것.
# how = '' / join 방식 : inner(left or right) - 교집합, outer - 합집합

a = pd.merge(a_1, a_2, on='자치구', how='inner')  # left join
a
```

```python
c = pd.merge(a, b, on='자치구', how='inner')
c
```
