# 소현

```python
# 소현
# scaler.csv

## EDA를 통해 결정한 내용
# - 문화시설 / 상업시설 / 유흥시설 : 시설 수의 변화는 근 3년간 거의 없는 것으로 판단되어 합산 후 하나의 컬럼으로 단일화 진행.
# (영화관수+공공공연장+민간공연장 → 문화시설수)
# (교육 서비스업+도매 및 소매업+숙박 및 음식점업+공공행정 국방 및 사회보장 행정 → 상업시설수)
# (예술 스포츠 및 여가관련 서비스업+단란주점 수 →유흥시설수) 최종 선정

# 데이터 기간
# 문화시설수 : 2018년
# 상업시설 : 2018년
# 유흥시설수 : 2018년
# 인구 : 2018년

# 2021.04.05 기준
# 유흥시설을 상업시설과 합침.

soso = pd.read_csv('dataset_문화_상업.csv')
soso
```

```python
soso.columns
```

```python
soso_old = pd.read_csv('./new_scaler.csv', index_col=0)
soso_old

# soso_old_old의 유흥시설수에서 술집 같은 단란주점을 제외시켜서
# soso_old의 유흥시설수를 보면 다소 감소한 모습을 볼 수 있음.
# soso에서는 단란주점을 제외시킨 유흥시설수와 상업시설수를 합친 것임.
```

```python
soso_old_old = pd.read_csv('./scaler.csv', index_col=0)
soso_old_old
```

```python
soso['1인당_문화시설수'] = soso['문화시설수'] / soso_old['인구']
soso['1인당_상업시설수'] = soso['상업시설수'] / soso_old['인구']
```

```python
soso
```

```python
soso.columns
```

```python
soso_scaled = soso[['자치구', '1인당_문화시설수', '1인당_상업시설수']]
soso_scaled
```

```python
x_train = soso_scaled.iloc[:,1:]


from sklearn.preprocessing import MinMaxScaler

MinMaxScaler = MinMaxScaler()

x_scaled = MinMaxScaler.fit_transform(x_train)

x_scaled
```

```python
df_x_scaled = pd.DataFrame(x_scaled, columns=['1인당_문화시설수_scaled', '1인당_상업시설수_scaled'])
df_x_scaled
```

```python
soso_scaled = pd.concat([soso_scaled['자치구'], df_x_scaled], axis=1)
soso_scaled
```

```python
soso_scaled.to_csv('soso_scaled.csv')
```
